# Check if all variables share the same timestamps

```python
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv("Datasets/DailyDelhiClimateTrain.csv")
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
```

```python
# Check for missing timestamps in any column
missing_by_column = df.isna().sum()
print("Missing values per variable:")
print(missing_by_column)
if missing_by_column.sum() == 0:
    print("All variables are synchronous and share the same timestamps.")
```
