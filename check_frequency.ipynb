# Check if the data has the same frequency

```python
import pandas as pd
```

```python
# Load dataset
df = pd.read_csv("Datasets/DailyDelhiClimateTrain.csv")
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
```

```python
# Calculate the difference between consecutive timestamps
freq_diff = df.index.to_series().diff().value_counts()
print("Frequency differences between timestamps:")
print(freq_diff)
if len(freq_diff) == 1 and freq_diff.index[0] == pd.Timedelta(days=1):
    print("The dataset has continuous measurements with a consistent daily frequency.")
```
